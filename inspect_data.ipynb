```python
import pandas as pd
```

```python
df = pd.read_csv("./audio_info.csv", index_col=0)
```

```python
df
```

```python
df[(df['source'] == 'McGill')]['gender'].value_counts()
```

```python
vox_df = df[df['source'] == 'VoxForge']
vox_group = vox_df.groupby('gender')
vox_sampled = vox_group.apply(lambda x: x.sample(vox_group.size().min() + (0 if (x['gender'] == 'female').all() else 58)).reset_index(drop=True))
```

```python
vox_final = vox_sampled.reset_index(0, drop=True)
```

```python
mcgill_df = df[df['source'] == 'McGill']
```

```python
final_df = pd.concat((vox_final, mcgill_df))
```

```python
final_df.to_csv('./audio_info_final.csv')
```
